# Creating a Machine Learning Model For Iris flower.

```python
#We use the skl-learn Library to make the Model and also for the data of iris.

from sklearn.datasets import load_iris
iris = load_iris()
```

# Getting the Data for input and Output.

```python
#In Machine Learning the 'x' denotes the Inputs in the model whereas 'y' denotes the target or the required output.

X = iris.data   #data input
y = iris.target #data output


#Storing the features of the iris flower.
feature_names = iris.feature_names
target_names = iris.target_names

#Both the x,y are of type "numpy.ndarray"
```

# Feature Names: 
1. sepal length (cm)
2. sepal width (cm)
3. petal length (cm)
4. petal width (cm)

# Target Names:
1. setosa
2. versicolor
3. virginica

#  Splitting the data into training and testing data.

```python
#The "train_test_split", Split arrays or matrices into random train and test subsets.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)  #Spliting the data 80% (input) and 20% (test).


#It prints the 'input' and 'test' size.
print(X_train.shape)  #(120, 4) is the dimension of the input array.
print(X_test.shape)   #(30, 4) is the dimension of the test array.

#We should have more train data than test data.
```

# Creating The 'Model' for the "Iris Flower".
# We use the "Nearest Neighbors Classification" Algorithm.

```python
#We now import from sklearn different alogrithms to create our model.
#We use the "Nearest Neighbors Classification" Algorithm for this model.

from sklearn.neighbors import KNeighborsClassifier


#We Create the model and speifiy the number of data points it need to observer.
#This is our Model.
knn = KNeighborsClassifier(n_neighbors = 3)

#We fitted and created the model. But we haven't checked the output.
knn.fit(X_train, y_train)

#We get the Predictions here.
y_pred = knn.predict(X_test)

```

# Comparing the Predictions to y_test values.

```python
#we now compare the predictions to the y_test values.

from sklearn import metrics

#It prints the model's accuracy.
print(metrics.accuracy_score(y_test, y_pred))
```

# Predicitng a new Iris Flower Photo inputted by a User.

```python
#Predicting a new iris flower photo inputted by a User.

#creating our new own sample data.
sample = [[3, 4, 4, 2], [2, 3, 5, 4]]

#storing the output in a predictions variable given to knn.predict model.
predictions = knn.predict(sample)

#predicitng the species.
pred_species = [iris.target_names[p] for p in predictions]  #list comprehension.

#Prints it's predictions it thinks is the species of the flower.
print("predictions: ", pred_species)  
```

# Saving the Model for use in the Future.

```python
#We import dump to save our model and laod to load our model.

from joblib import dump, load

#This stores our model in a binary file.
dump(knn, 'iris flower predict.joblib')
```

# Importing the Model from a Saved file.

```python
#Now we can import the saved model and run without having to train it multiple times.

#Loading the model
model = load('iris flower predict.joblib')

#Running the Model again.
model.predict(X_test)


#Running the saved model with 'sample data'.
sample = [[3, 4, 4, 2], [2, 3, 5, 4]]

predictions = model.predict(sample)
pred_species = [iris.target_names[p] for p in predictions]

print("predictions: ", pred_species)  
```
