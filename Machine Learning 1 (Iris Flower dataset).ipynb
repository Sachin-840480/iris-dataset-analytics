# Creating a Machine Learning Model For Iris flower.

```python
#We use the skl-learn Library to make the Model and also for the data of iris.

from sklearn.datasets import load_iris
iris = load_iris()
```

# Getting the Data for input and Output.

```python
#In Machine Learning the 'x' denotes the Inputs in the model whereas 'y' denotes the target or the required output.

X = iris.data   #data input
y = iris.target #data output


#Storing the features of the iris flower.
feature_names = iris.feature_names
target_names = iris.target_names

#Both the x,y are of type "numpy.ndarray"
```

# Feature Names: 
1. sepal length (cm)
2. sepal width (cm)
3. petal length (cm)
4. petal width (cm)

# Target Names:
1. setosa
2. versicolor
3. virginica

# Splitting the data into training and testing data.

```python
#The "train_test_split", Split arrays or matrices into random train and test subsets.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)  #Spliting the data 80% (input) and 20% (test).

print(X_train.shape)  #(90, 4) is the dimension of the array.
print(X_test.shape)   #(60, 4) is the dimension of the array.

#We should have more train data than test data.
```
